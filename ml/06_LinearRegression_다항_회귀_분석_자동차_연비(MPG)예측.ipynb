문제 정의 : LinearRegression 다항 회귀 분석 사용하여 1970년대 후반과 1980년대 초반의 자동차 연비(MPG)를 예측

```python
# 기본 라이브러리 불러오기
import pandas as pd
import numpy as np
import seaborn as sns

#한글 깨짐 방지
import matplotlib as mpl
import matplotlib.pyplot as plt
 
%config InlineBackend.figure_format = 'retina'
 
!apt -qq -y install fonts-nanum
 
import matplotlib.font_manager as fm

fontpath = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf'
font = fm.FontProperties(fname=fontpath, size=9)
plt.rc('font', family='NanumBarunGothic') 
mpl.font_manager._rebuild()
```

### step 1) 데이터 셋 준비하기

```python
# csv파일을 데이터프레임으로 변환
df = pd.read_csv('auto-mpg.csv',header=None)
# 열 이름 지정
df.columns = ['mpg','cylinders','displacement','horsepower','weight',
              'acceleration','model year','origin','name']
# 데이터 살펴보기
df.head()
```

### step 2) 데이터 탐색 및 데이터 전처리

```python
# 데이터 자료형 확인
df.info()
```

```python
# horsepower 열의 자료형 변경 (문자열-> 숫자)
df['horsepower'].unique()
```

```python
# '?'를 np.nan으로 변경
df['horsepower'].replace('?',np.nan,inplace=True)
df.info()
```

```python
# 누락 데이터를 삭제
df.dropna(subset=['horsepower'],axis=0,inplace=True)
df.info()
```

```python
# 문자열 -> 숫자(실수타입)로 변환
df['horsepower'] = df['horsepower'].astype('float')
df.info()
```

### step 3) 속성(feature 또는 variable) 선택

```python
#분석에 활용할 속성을 선택 (연비, 실린더, 마력, 중량)
ndf = df[['mpg','cylinders','horsepower','weight']]
ndf.head()
```

### step 4) 데이터셋 구분 - 훈련용/검증용

```python
# 속성(변수) 선택 -> 다항 회귀
X = ndf[['cylinders','horsepower','weight']] # 독립변수 X1,X2,X3
y = ndf['mpg'] # 종속변수 y
```

```python
# seaborn pairplot으로 두 변수간의 모든 경우의 수 그리기
sns.pairplot(ndf)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, #독립변수
                                                 y, #종속변수
                                                 test_size=0.3, #검증 30%
                                                 random_state=10) #랜덤추출값
```

### step 5) 다항 회귀분석 모형 - sklearn 사용

```python
# sklearn 라이브러리에서 선형회귀분석 모듈 가져오기
from sklearn.linear_model import LinearRegression

# 다항 회귀분석 모형 객체 생성 후 학습 시키기
lr = LinearRegression().fit(X_train,y_train)
```

### step 6) 성능 평가하기

```python
# 학습을 마친 모형에 test data를 적용하여 결정계수 (R-제곱) 계산
# -> 결정계수 값이 클수록 모형의 예측 능력이 좋다고 판단
r_square = lr.score(X_test,y_test)
r_square
```

```python
# 모형에 전체 X 데이터를 입력하여 예측한 값 y_hat으로 담기
y_hat = lr.predict(X)
# 모형에 테스트 X_test 데이터를 입력하여 예측한 값 y_hat_test으로 담기
y_hat_test = lr.predict(X_test)
# train 데이터(X_train)의 산점도와 test 데이터로 예측한 (y_hat_test)회귀선을 그래프로 출력
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(1,1,1)
# 데이터의 분포
ax.plot(X_train,y_train,'o',label='Train Data')
ax.plot(X_test,y_hat_test,'r+',label='Predicted Value')
ax.legend(loc='best')
plt.xlabel('weight')
plt.ylabel('mpg')
plt.show()
```

```python
print("훈련 세트 정확도 : {:.2f}%".format(lr.score(X_train,y_train)*100))
print("테스트 세트 정확도 : {:.2f}%".format(lr.score(X_test,y_test)*100))
```

```python
plt.figure(figsize=(10,5))
ax1 = sns.kdeplot(y,label='y')
ax2 = sns.kdeplot(y_hat, label='y_hat',ax=ax1)
plt.title('LinearRegression 다항 회귀 분석 자동차 연비(MPG) 예측')
plt.legend()
plt.show()
```
