### 신도림 데이터 확인하기

in 연산자를 활용해 우리 동네 이름이 포함된 지역명 찾기

```python
print('신도림' in '서울특별시 구로고 신도림동(1153051000)')
print('1153' in '서울특별시 구로고 신도림동(1153051000)')
print('()' in '서울특별시 구로고 신도림동(1153051000)')
```

in 연산자를 활용해 신도림의 데이터 출력하기

```python
import csv
f = open('age.csv', encoding = 'cp949')
data = csv.reader(f)

for row in data:
  if '신도림' in row[0]:
    print(row)
```

연령 별 인구수(0-100세 이상) 데이터 확인 -> 4번

```python
import csv
f = open('age.csv', encoding = 'cp949')
data = csv.reader(f)

for row in data:
  if '신도림' in row[0] :
    print(row[3:])
    print(len(row[3:]))
```

### 신도림 연령별 인구수 result 리스트에 저장하기 -> 시각화를 위해

```python
import csv
f = open('age.csv', encoding = 'cp949')
data = csv.reader(f)

result=[]

for row in data:
  if '신도림' in row[0]:
    for i in row[3:]:
      result.append(int(i)) #인구 수 데이터 정수로 반환 후 result 변수에 담기

print(result)
```

신도림 연령 별 인구수 데이터 시각화 하기

```python
import matplotlib.pyplot as plt

plt.figure(dpi=200)
plt.style.use('ggplot') #격자 무늬 스타일로 지정
plt.plot(result)
```
